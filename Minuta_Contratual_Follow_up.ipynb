```python
import pandas as pd
from google.colab import files
import datetime

hoje = datetime.datetime.now()
```

```python
nomeBase = input("Digite exatamente o nome do arquivo base e a sua extensão (ex: base.xlsx): ") #Follow up Minuta Contratual.xlsx
minutasBase = pd.read_excel(nomeBase, sheet_name=0, skiprows=4, index_col=None, usecols= 'A, B, E, J, M, Q, V, AA, AD, AF, AG, AH, AI') #, header=None, names = colunasNome)
minutasBase.columns = minutasBase.columns.str.replace('\n','_')
#display(minutasBase)
```

```python
dfMinutasAll = pd.DataFrame(minutasBase)
#display(dfMinutasAll)
dfMinutas = dfMinutasAll.dropna(subset=["#"])
display(dfMinutas)
```

```python
minutasConcluido = dfMinutas.loc[(dfMinutas["STATUS"] == "Concluído")].sort_values(by=["#"])
minutasAndamento = dfMinutas.loc[(dfMinutas["STATUS"] == "Em Andamento")].sort_values(by=["#"])
```

```python
# CONCLUIDO
qntMinuta = minutasConcluido["#"].count()
emissaoMinuta = minutasConcluido["SLA envio de minuta"].mean()
primAjuste = minutasConcluido["SLA 1º ajuste"].mean()
segAjuste = minutasConcluido["SLA 2º ajuste"].mean()
tercAjuste = minutasConcluido["SLA 3º ajuste"].mean()
discussasoMinuta = minutasConcluido["SLA DISCUSSÃO MINUTA"].mean()
mediaAssinatura = minutasConcluido["SLA assinatura"].mean()
mediaTotal = minutasConcluido["SLA total"].mean()
totalNegociado = minutasConcluido["VALOR DO CONTRATO_ou VALOR ADITADO (ñ acumulado)"].sum()
leadTime = dfMinutas["SLA CONCLUSÃO _PROCESSO_SUPRIMENTOS"].mean()

# EM ANDAMENTO
qntMinutaA = len(minutasAndamento["#"])
qntFornecedor = len(minutasAndamento.loc[(minutasAndamento["ETAPA"] == "Fornecedor")])
qntSuprimentos = len(minutasAndamento.loc[(minutasAndamento["ETAPA"] == "Suprimentos")])
qntAguardando = len(minutasAndamento.loc[(minutasAndamento["ETAPA"] == "Aguardando envio de minuta")])
primAjuste_A = len(minutasAndamento.loc[(minutasAndamento["ETAPA"] == "Aguardando minuta 1º ajuste")])
segAjuste_A = len(minutasAndamento.loc[(minutasAndamento["ETAPA"] == "Aguardando minuta 2º ajuste")])
tercAjuste_A = len(minutasAndamento.loc[(minutasAndamento["ETAPA"] == "Aguardando minuta 3º ajuste")])
qntClicksign = len(minutasAndamento.loc[(minutasAndamento["ETAPA"] == "Aguardando envio ClickSign")])
qntAssinatura = len(minutasAndamento.loc[(minutasAndamento["ETAPA"] == "Em assinatura")])
totalNegociado_A = round(minutasAndamento["VALOR DO CONTRATO_ou VALOR ADITADO (ñ acumulado)"].sum(), 2)
totalEmAndamento = round(minutasAndamento["SLA TOTAL_EM ANDAMENTO_HOJE"].mean(), 1)
```

```python
# =========== REPORT =========== #
print("\tCONCLUIDO\n")
print(f"Qnt (Concluído): {qntMinuta}")
print(f"Média Emissão de Minuta: {round(emissaoMinuta, 1)}")
print(f"Média 1º Ajuste: {round(primAjuste, 1)}")
print(f"Média 2º Ajuste: {round(segAjuste, 1)}")
print(f"Média 3º Ajuste: {round(tercAjuste, 1)}")
print(f"Média Discussão Minuta: {round(discussasoMinuta, 1)}")
print(f"Média Assinatura: {round(mediaAssinatura, 1)}")
print(f"Média Total: {round(mediaTotal, 1)}")
print(f"Total Negociado: {round(totalNegociado, 2)}")
print(f"Lead Time Compras: {round(leadTime, 1)}")
print("\n\n")
# ===== ===== #
print("\tEM ANDAMENTO\n")
print(f"Qnt (Em andamento): {qntMinutaA}")
print(f"Etapa - Fornecedor: {qntFornecedor}")
print(f"Etapa - Suprimentos: {qntSuprimentos}")
print(f"Etapa - Aguardando envio de minuta: {qntAguardando}")
print(f"Etapa - Aguardando minuta 1º ajuste: {primAjuste_A}")
print(f"Etapa - Aguardando minuta 2º ajuste: {segAjuste_A}")
print(f"Etapa - Aguardando minuta 3º ajuste: {tercAjuste_A}")
print(f"Etapa - Aguardando envio ClickSign: {qntClicksign}")
print(f"Etapa - Em assinatura: {qntAssinatura}")
print(f"Total Negociado: {totalNegociado_A}")
print(f"SLA Total Em Andamento: {totalEmAndamento}") 
```

```python
# =========== PARA MONTAR O GRAFICO =========== #
graph1_data = {"Etapas": ["Lead Time Compras", "Emissão Minuta", "Discussão Minuta (forn, supr, jur)", "Assinatura de Minuta", "Contrato Jurídico"], "Médias": [round(leadTime, 1), round(emissaoMinuta, 1), round(discussasoMinuta, 1), round(mediaAssinatura, 1), round(mediaTotal, 1)]}
graph1_df = pd.DataFrame.from_dict(graph1_data)
display(graph1_df)
print("\n")
somaMinutas = (qntMinuta+qntMinutaA)
graph2_data = {"Estado": ["Concluído", "Em andamento", "Total"], "Valores": [qntMinuta, qntMinutaA, somaMinutas], "Porcentagem": [("{:.1%}".format(qntMinuta/somaMinutas)), ("{:.1%}".format(qntMinutaA/somaMinutas)), ("{:.0%}".format(somaMinutas/somaMinutas))]}
graph2_df = pd.DataFrame.from_dict(graph2_data)
display(graph2_df)
print("\n")
graph3_data = {"Status": ["SLA TOTAL", "SLA EM ANDAMENTO"], "Valores": [round(mediaTotal, 1), round(totalEmAndamento, 1)]}
graph3_df = pd.DataFrame.from_dict(graph3_data)
display(graph3_df)

# ========== EXPORTANDO PRO EXCEL ========== #
#titulo = f'{hoje.month}_{hoje.year}_Minuta Contratual.xlsx'
# baixarExcel = resultFinal4.to_excel(titulo)
# files.download(titulo)
```
