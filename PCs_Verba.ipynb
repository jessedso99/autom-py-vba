```python
import pandas as pd
from google.colab import files
import datetime

hoje = datetime.datetime.now()
```

```python
nomeBase = input("Digite exatamente o nome do arquivo base e a sua extensão: (ex: base.xlsx)")
```

```python
pcsdata = pd.read_excel(nomeBase, sheet_name=0) #, index_col=False) 
```

```python
dfdata = pd.DataFrame(pcsdata, columns =["PC", "FORNECEDOR", "RAZAO_SOC", "MÊS EMISSÃO", "DESC_CCUST", "VLR_TOTAL", "ANO", "PRC_UNIT"])
dfdata["PC"] = dfdata["PC"].astype("object")
```

```python
filtros1 = (dfdata.loc[(dfdata["ANO"] == hoje.year) & (dfdata["MÊS EMISSÃO"] == (hoje.month-1)) & (dfdata["PRC_UNIT"] == 1)]).sort_values(by=["PC"])
filtros1Soma = filtros1.groupby(["PC", "FORNECEDOR", "RAZAO_SOC", "DESC_CCUST"], as_index=False)["VLR_TOTAL"].sum()
filtros1Soma = filtros1Soma.sort_values(by=["PC"])
filtros1Soma["REAL_VLR_TOTAL"] = ""
valSuperior = filtros1Soma.loc[(filtros1Soma["VLR_TOTAL"] >= 100000)]
display(valSuperior)
```

```python
listaPCs = valSuperior.loc[:,"PC"].values.tolist()
new_df = valSuperior.values.tolist()
```

```python
for i in range(0, len(new_df)):
  test_list = filtros1.loc[(filtros1["PC"] == listaPCs[i])]
  new_df[i][5] = sum(list(set(test_list["VLR_TOTAL"])))
#display(new_df)
```

```python
resultado = pd.DataFrame(new_df).drop(columns=4)
resultado.rename(columns={0: "PC", 1: "FORNECEDOR", 2: "RAZAO_SOC", 3: "DESC_CCUST", 5: "REAL_VLR_TOTAL"}, inplace=True)

resultFinal = resultado.loc[(resultado["REAL_VLR_TOTAL"] >= 100000)]
display(resultFinal)
print("\n\n")
resultFinal2 = resultFinal.groupby('DESC_CCUST').agg({"REAL_VLR_TOTAL":["count", "sum"]})
display(resultFinal2)
```

```python
#SALVAR OS RESULTS
titulo = f'{hoje.month}_{hoje.year}_PCs_Verba.xlsx'
baixarExcel = resultFinal.to_excel(titulo, index=False)
files.download(titulo)
```
