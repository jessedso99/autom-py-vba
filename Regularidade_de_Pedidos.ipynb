```python
import pandas as pd
from google.colab import files
import datetime
import numpy as np
import calendar

hoje = datetime.datetime.now()
```

```python
nomeBase = input("Digite exatamente o nome do arquivo base e a sua extensão: (ex: base.xlsx)") #"2023_2_BRACELL.xlsx" # testee base.xlsx)ITAQUÍ.xlsx
```

```python
####################### IMPORTANTE #######################
# Verficar o nome da Sheet que contém os dadaos da base.
# Verificar se é necessário SKIP ROWS
# Verificar nome das Colunas "DT Emissao" e "DT Emissao.1"

bracellBase = pd.read_excel(nomeBase, sheet_name="1-Pedidos de Compra", skiprows=1) ####, index_col=False)
#dfBracell = pd.DataFrame(bracellBase, columns =["Documento", "Razao Social", "DT Emissao PC", "DT Emissao NF", "Unidade"])
dfBracell = pd.DataFrame(bracellBase, columns =["Documento", "Razao Social", "DT Emissao.1", "DT Emissao", "Unidade"])

#display(dfBracell)#.dtypes)
```

```python
#bracellTable = dfBracell.groupby(["Documento", "Razao Social", "DT Emissao PC", "DT Emissao NF"], as_index=False)["Unidade"].count()
bracellTable = dfBracell.groupby(["Documento", "Razao Social", "DT Emissao.1", "DT Emissao"], as_index=False)["Unidade"].count()
bracellTable = pd.DataFrame(bracellTable).drop(columns="Unidade") #bracellTable.rename(columns={"Unidade": "Quantidade Produtos"}, inplace=True)
#display(bracellTable)
```

```python
### ADD AND CALCULATE COLUMN DIFERENÇA
bracellTable["DIFERENÇA"] = (bracellTable["DT Emissao.1"] - bracellTable["DT Emissao"]) / np.timedelta64(1, "D")
#bracellTable["DIFERENÇA"] = (bracellTable["DT Emissao PC"] - bracellTable["DT Emissao NF"]) / np.timedelta64(1, "D") #.map(lambda x: x[:-5])# / np.timedelta64(1, "D")
#display(bracellTable)

### ADD AND CALCULATE COLUMN REGULARIDADE
regularidadeCalc = []
for linha in bracellTable["DIFERENÇA"]:
  if linha > 0 : regularidadeCalc.append("IRREGULAR")
  else: regularidadeCalc.append("REGULAR")
bracellTable["REGULARIDADE"] = regularidadeCalc

display(bracellTable) #pd.DataFrame(bracellTable, columns = ["Documento"]))
```

```python
resultado1 = bracellTable.groupby("REGULARIDADE")["REGULARIDADE"].count()
result_edit = (dict(resultado1))

regularPercent = 0 if "REGULAR" not in result_edit else (result_edit["REGULAR"]/sum(result_edit.values()))
#regularPercent = result_edit["REGULAR"]/sum(result_edit.values())
irregularPercent = 0 if "IRREGULAR" not in result_edit else (result_edit["IRREGULAR"]/sum(result_edit.values()))
#irregularPercent = result_edit["IRREGULAR"]/sum(result_edit.values())

############ Resultado Final ############
print(f'{nomeBase[:-5].upper()}\n{calendar.month_name[hoje.month]}')
print(pd.DataFrame(resultado1))
print(f'Total                  {sum(bracellTable.groupby("REGULARIDADE")["REGULARIDADE"].count())}')
#display([{"IRREGULAR": "{:.0%}".format(irregularPercent)}, {"REGULAR": "{:.0%}".format(regularPercent)}])
print("IRREGULAR {:.0%}  REGULAR {:.0%}".format(irregularPercent, regularPercent))
```
