```python
import pandas as pd
import numpy as np
from google.colab import files
import datetime
```

```python
nomeBase = input("Digite exatamente o nome do arquivo base e a sua extensão: (ex: base.xls)") #"relatorio_chamados_concluidos.csv"
#base = pd.read_csv(nomeBase, sep=";", encoding="utf_8", quoting=csv.QUOTE_ALL)
base = pd.read_excel(nomeBase, skiprows=12, usecols= 'A:O') #Q
#display(base)
#print(np.shape(base))
dfBase = pd.DataFrame(base)
#display(dfBase)
```

```python
colTempoVida = dfBase["Tempo de Vida"]#pd.DataFrame(dfBase["Tempo de Vida"])

spliTempoVida = []
tempoVida = []
for i in range(0, len(colTempoVida)):
  spliTempoVida.append(colTempoVida[i].split(sep= " "))
  if (spliTempoVida[i][0][-1]) != "d":
    tempoVida.append("1D")
  elif (spliTempoVida[i][0][-1]) == "d":
    if (int(spliTempoVida[i][0][:-1])) >= 5:
      tempoVida.append("5D+")
    elif (int(spliTempoVida[i][0][:-1])) == 1:
      tempoVida.append("1D")
    elif (int(spliTempoVida[i][0][:-1])) == 2:
      tempoVida.append("2D")
    elif (int(spliTempoVida[i][0][:-1])) == 3:
      tempoVida.append("3D")
    elif (int(spliTempoVida[i][0][:-1])) == 4:
      tempoVida.append("4D")

colDias = pd.DataFrame(tempoVida)
# display(colDias)
# len(tempoVida)
```

```python
#Add columns DIAS
dfBase["DIAS"] = colDias

tipoChamado = []
for linha in dfBase["Categoria"]:
  if linha == "Liberação de fornecedor" : tipoChamado.append("FORNECEDORES")
  elif linha == "Liberação de fornecedor e cadastro de dados bancários" : tipoChamado.append("FORNECEDORES")
  elif linha == "Outros assuntos relacionados a fornecedores" : tipoChamado.append("FORNECEDORES")
  elif linha == "Cadastro de dados bancários" : tipoChamado.append("FORNECEDORES")
  elif linha == "Cadastro de novo produto" : tipoChamado.append("PRODUTOS")
  elif linha == "Outros assuntos relacionados a produtos" : tipoChamado.append("PRODUTOS")
  elif linha == "Alteração de produto" : tipoChamado.append("PRODUTOS")
  else: tipoChamado.append("### NOVA CATEGORIA ###")
dfBase["TIPO CHAMADO"] = tipoChamado

display(dfBase)
```

```python
tipoFornecedor = dfBase.loc[(dfBase["TIPO CHAMADO"] == "FORNECEDORES")]
resultFornec = pd.DataFrame(tipoFornecedor.groupby("DIAS")["DIAS"].count())
display("FORNECEDOR", resultFornec)
```

```python
tipoProduto = dfBase.loc[(dfBase["TIPO CHAMADO"] == "PRODUTOS")]
resultProd = pd.DataFrame(tipoProduto.groupby("DIAS")["DIAS"].count())
display("PRODUTOS", resultProd)
```

```python
#SALVAR OS RESULTS
hoje = datetime.datetime.now()
titulo = f'{hoje.month}_{hoje.year}_SLA Cadastro de Produto.xlsx'
baixarExcel = dfBase.to_excel(titulo, index=False)
#baixarExcel = resultFinal2.to_excel(titulo)#, index=False)
files.download(titulo)
```
